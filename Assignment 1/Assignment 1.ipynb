```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def cost_plotting(n, k):
    x = np.array(list(range(1, n+1)))
    theta_logn = np.log2(x)
    theta_n = x
    theta_n2 = x**2
    theta_n3 = x**3
    theta_nk = x**k
    fig, ax = plt.subplots()
    
    ax.plot(x, theta_logn, label='theta(logn)', marker='o')
    ax.plot(x, theta_n, label='theta(n)', marker='o')
    ax.plot(x, theta_n2, label='theta(n^2)', marker='o')
    ax.plot(x, theta_n3, label='theta(n^3)', marker='o')
    ax.plot(x, theta_nk, label='theta(n^k)', marker='o')
    ax.axis(xmin=0)
    ax.set_xlabel('input size (n)')
    ax.set_ylabel('computation cost (theta)')
    leg = ax.legend();
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200 
cost_plotting(7, 4)
```
